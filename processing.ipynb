```python
import pandas as pd
import numpy as np
```

```python
fake=pd.read_csv("Fake.csv")
true=pd.read_csv("True.csv")
```

```python
fake.head()
```

```python
true.head()
```

```python
fake["label"]=0
true["label"]=1
```

```python
fake.head()
```

```python
true.head()
```

```python
df = pd.concat([fake, true], ignore_index=True)
```

```python
df["content"] = df["title"] + " " + df["text"]
```

```python
x=df["content"] 
y=df["label"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
feature_extraction = TfidfVectorizer(
    min_df=1,
    stop_words='english',
    lowercase=True
)
```

```python
x_train_fe=feature_extraction.fit_transform(x_train)
x_test_fe=feature_extraction.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train_fe,y_train)
```

```python
test_predection=model.predict(x_train_fe)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy =", accuracy_score(y_train,test_predection))
```

```python
import joblib
```

```python
joblib.dump(model,"fake_news_model2.pkl")
```

```python
joblib.dump(feature_extraction,"vector_model2.pkl")
```
